```python
i=0
while(i<999):   
    if(rainfall[i][1]==2000):    
      kharifmean=0
      rabimean=0
      summermean=0
      wholemean=0
      while(i<999 and rainfall[i][1]<2015):
        kharifmean=(rainfall[i][8]+rainfall[i][9]+rainfall[i][10]+rainfall[i][11])/4
        rabimean=(rainfall[i][2]+rainfall[i][3]+rainfall[i][4]+rainfall[i][12]+rainfall[i][13])/5
        summermean=(rainfall[i][5]+rainfall[i][6]+rainfall[i][7])/3
        for j in range(2,14):
          wholemean+=rainfall[i][j]
        wholemean=wholemean/12
        district_name=rainfall[i][0]
        print(i)
        raindict[str(district_name)][rainfall[i][1]]['Kharif']=kharifmean
        raindict[str(district_name)][rainfall[i][1]]['Rabi']=rabimean
        raindict[str(district_name)][rainfall[i][1]]['Summer']=summermean
        raindict[str(district_name)][rainfall[i][1]]['Whole Year']=wholemean
        i+=1        
    else:
        i+=1
```
